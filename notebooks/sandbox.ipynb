```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import FormatStrFormatter
```

```python
steps_data = pd.read_csv("C:/Users/davis.pazars/Documents/orkans/results/nowcasts_steps.csv")
sseps_data = pd.read_csv("C:/Users/davis.pazars/Documents/orkans/results/nowcasts_sseps.csv")
```

```python
steps_data_sorted = steps_data.loc[:, ['n_ens_members', 'roc_area_T15_THR0.1', 'nwc_run_time']].sort_values(by='n_ens_members', ascending=True)
sseps_data_sorted = sseps_data.loc[:, ['n_ens_members', 'roc_area_T15_THR0.1', 'nwc_run_time']].sort_values(by='n_ens_members', ascending=True)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(12, 5)

ax1.plot(steps_data_sorted['n_ens_members'], steps_data_sorted['roc_area_T15_THR0.1'] * 100, ".-", markersize=10, label="STEPS")
ax1.plot(sseps_data_sorted['n_ens_members'], sseps_data_sorted['roc_area_T15_THR0.1'] * 100, ".--", markersize=10, label="SSEPS")
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
ax1.set_xlabel("Ensemble members")
ax1.set_ylabel("Area under ROC curve, %")
ax1.legend()

ax2.plot(steps_data_sorted['n_ens_members'], steps_data_sorted['nwc_run_time'], ".-", markersize=10, label="STEPS")
ax2.plot(sseps_data_sorted['n_ens_members'], sseps_data_sorted['nwc_run_time'], ".--", markersize=10, label="SSEPS")
ax2.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
ax2.set_xlabel("Ensemble members")
ax2.set_ylabel("Calculation time, s")
ax2.legend()

plt.show()
```
